```python
#Importing the required libraries
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#Importing the csv file and only selecting "Class" and "ProcessedListing" column
df = pd.read_csv("Final_Processed.csv",usecols=["Class","ProcessedListings"])
df
```

```python
#selecting feautures in X 
X = df.iloc[:,1].values
X
```

```python
#storing the dependent variable in y
y = df.iloc[:,0].values
y
```

```python
#Creating a tfidf vectorizer to transform processed listing into maxtrix form with minimum document frequency of 10 and including unigram and bigram words
tfidf = TfidfVectorizer(min_df=10,ngram_range=(1,2))
tfidf.fit(X)
```

```python
#Transforming the features in matrix form
X_transformed = tfidf.transform(X)
X_transformed = X_transformed.toarray()
X_transformed
```

```python
#from sklearn.compose import ColumnTransformer
#from sklearn.preprocessing import OneHotEncoder
#ct = ColumnTransformer(transformers=[("encoding",OneHotEncoder())],remainder = "passthrough")
```

```python
y
```

```python
#Tranforming the dependent variables into "One Hot" encoded form
#Label encoding the dependent variables and then tranforming it to one hot form using onehotencoder
le = LabelEncoder()
y_transformed = le.fit_transform(y)
onehotencoder = OneHotEncoder()
y_transformed = y_transformed.reshape(-1, 1)
y_transformed = onehotencoder.fit_transform(y_transformed).toarray()
y_transformed
```

```python
#Splitting the data into traning set and test set 
x_train,x_test,y_train,y_test = train_test_split(X_transformed,y_transformed,test_size=0.2,random_state=1, stratify=y_transformed)
```

```python
#Creting a sequential model
ann = tf.keras.models.Sequential()
#Input Layer with the number of neurons equal to the lenght of tensor in X_transformed
ann.add(tf.keras.layers.Dense(units = 3234, activation = "relu"))
#1st Hidden Layers with 6 neurons and relu activation function
ann.add(tf.keras.layers.Dense(units = 6, activation = "relu"))
#2nd Hidden Layers with 6 neurons and relu activation function
ann.add(tf.keras.layers.Dense(units = 6, activation = "relu"))
#Adding Dropout layer to reduce the chances of overfitting
ann.add(tf.keras.layers.Dropout(0.25))
#3rd Hidden Layers with 6 neurons and relu activation function
ann.add(tf.keras.layers.Dense(units = 6, activation = "relu"))
#4th Hidden Layers with 6 neurons and relu activation function
ann.add(tf.keras.layers.Dense(units = 6, activation = "relu"))
#Output Layer with 9 neurons as we have 9 distinct categories in the dependent variable y
#Activation function is softmax as we are dealing with multiclass classification problem
ann.add(tf.keras.layers.Dense(units = 9, activation = "softmax"))

#Compiling the ann model with adam optimizer and loss function as categorical_crossentropy as its a multiclass(more than 2 classes) classification
ann.compile(optimizer = "adam",loss = "categorical_crossentropy", metrics= ["accuracy"] )
#Training the ANN with train data with a batch size of 32 and epochs 100
ann.fit(x_train,y_train,epochs=100,batch_size = 32)
#selecting the index position of the class with value 1 in the test data.
y_test_arg=np.argmax(y_test,axis=1)
#Making Predictions and selecting the index position of class with highest probability
Y_pred = np.argmax(ann.predict(x_test),axis=1)
```

```python
#Printing the results using confusion matrix and accuracy score
print('Confusion Matrix')
print(confusion_matrix(y_test_arg, Y_pred))
print('Accuracy Score')
print(accuracy_score(y_test_arg, Y_pred))
```
