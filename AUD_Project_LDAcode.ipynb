```python
import pandas as pd
```

```python
reviewlist= pd.read_csv('final rawdata.csv')

reviewlist.head()
```

```python
reviewlist["FullReview"] = reviewlist["Title"]+" "+reviewlist["Description"]
```

```python
reviewlist.iloc[0].FullReview
```

```python
reviewsTitle = reviewlist["FullReview"].tolist()
reviewsDesc = reviewlist["Description"].tolist()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


titletokens = []
stop_word = stopwords.words("english")
stop_word.extend(["free","has","was","quot","gt","etc","bk","brookville","borough","lt","affordable","excellent",
                      "guarantee","midtown","ct","ny","long","brouoghts","boroughs","gloria","borough","astoria","rockland",
                  "fairfield","staten","bridgeport","bx","si","li","suffolk","ul","ol","weschester","soho","stamford",
                  "chelsea","nassau","county","connecticut","yonkers","west","nj","upper","east","williamsburg","br",
                  "harlem","new","york","westchester","brooklyn","manhattan","nyc","bklyn","bronx","queens","island",
                  "jersey","stratford","joshua","joes","apartment","available","boros","year","downtown","greenpoint"
                 ,"apt","hempstead","lower","village","today","ac","accept","accepted","according",
                 'able', 'accent', 'accurate', 'across', 'activity', 'actual', 'ad', 'add', 'added', 'addition', 'additional', 'address', 'adult', 'advance', 'agency', 'aging', 'agreement', 'aid', 'aide', 'air',
                 'alex', 'already', 'also', 'always', 'another', 'answer', 'anyday', 'anything', 'anytime', 'anywhere', 'apartment', 'app', 'apple',
                 'armonk',"de","por","en","la", 'around', 'arrangement', 'arrival', 'way', 'wear', 'weather', 'web', 'website', 'week', 'weekday', 'weekend', 'weekly', 'welcome','whatever', 'whatsapp', 'whether', 'white', 'whole', 'wide', 'willing','within', 'without', 'wo', 'woman','working', 'workload', 'worry', 'would', 'wow','written', 'xtreme', 'yard', "year", 'youtube', 'zelle', 'zip'])

for titles in reviewsTitle:
    filtered = []
    tokens = word_tokenize(titles)
    for w in tokens:
        if w.lower() not in stop_word and w.isalpha():
            filtered.append(w.lower())
    titletokens.append(filtered)

#print(titletokens)
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_title = []
for eachtoken in titletokens:
    #for token in eachtoken:
    #    token.lower()
    #    lemmatized_review.append(lemmatizer.lemmatize(token)
        
    lemmatized_title.append([lemmatizer.lemmatize(token) for token in eachtoken])
    
#print(lemmatized_title)
```

```python
newlist = []
for toks in lemmatized_title:
    newlist.append(" ".join(toks))

#print(newlist)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
vectorizer1 = TfidfVectorizer(ngram_range=(1,1),min_df=10)
vectorizer1.fit(newlist)
#you need to fit the vectorizer against the complete collection of words, and only then transform to get the count of the words
#print(vectorizer1.vocabulary_)
X = vectorizer1.transform(newlist).toarray()
terms = vectorizer1.get_feature_names()
print(terms)
```

```python
from sklearn.decomposition import LatentDirichletAllocation 
lda = LatentDirichletAllocation(n_components=10).fit(X)
```

```python
topicslist = []
for topic_idx, topic in enumerate(lda.components_): 
    templist = []
    templist.extend([topic_idx+1," ".join([terms[i] for i in topic.argsort()[:-12:-1]])])
    topicslist.append(templist)
    print("Topic %d:" % (topic_idx+1))
    print(" ".join([terms[i] for i in topic.argsort()[:-12:-1]]))
```

```python
import numpy as np
service_topics = lda.transform(X)
# column names
topicnames = ["Topic" + str(i+1) for i in range(15)]
# index names
docnames = ["Service " + str(i+1) for i in range(len(X))]
# Make the pandas dataframe
df_service_topic = pd.DataFrame(np.round(service_topics, 4), columns=topicnames, index=docnames)
# Get dominant topic for each document
dominant_topic = np.argmax(df_service_topic.values, axis=1)+1
df_service_topic['dominant_topic'] = dominant_topic
```

```python
print("The Dominant topics of top ten Restaurant Reviews are: ")
topicstotitle = []
for i in range(len(df_service_topic)):
    templist = []
    val = sorted(df_service_topic.iloc[i].values)[-2:-6:-1]
    print("Service :",i+1,"\t Topic:",(df_service_topic.iloc[i].tolist().index(val[0]))+1,
          "\t",(df_service_topic.iloc[i].tolist().index(val[1]))+1,
          "\t",(df_service_topic.iloc[i].tolist().index(val[2]))+1,
          "\t",(df_service_topic.iloc[i].tolist().index(val[3]))+1)
    templist.extend(["service :"+str(i+1),reviewsTitle[i],(df_service_topic.iloc[i].tolist().index(val[0]))+1,
                    (df_service_topic.iloc[i].tolist().index(val[1]))+1,(df_service_topic.iloc[i].tolist().index(val[2]))+1,
                    (df_service_topic.iloc[i].tolist().index(val[3]))+1])
    topicstotitle.append(templist)
```

```python
topics_df = pd.DataFrame(topicslist,columns=['Topic Number','Topic Weighted Words'])
```
