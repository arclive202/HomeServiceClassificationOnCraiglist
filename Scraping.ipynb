```python
# Setting up urls in Beautiful Soup

website = []
for i in range(0, 2000, 120):
    website.append("https://newyork.craigslist.org/d/household-services/search/hss?s=" + str(i))

urls = []
    
for j in website:
    headers = random.choice(headers_list)
    r = requests.Session()
    r.headers = headers
    
    html = r.get(j).text
    soup = BeautifulSoup(html)
    for k in soup.find_all('a', class_ = 'result-title hdrlnk'):
        urls.append(k['href'])
        
print(urls[0:5])
print(len(urls))
```

```python
# Getting title and text from urls

ls1 = []
ls2 = []
ls3 = []
ls4 = []
prefinal = []

title = []
text = []

for l in urls:
    headers = random.choice(headers_list)
    r = requests.Session()
    r.headers = headers
    
    html2 = r.get(l).text
    soup = BeautifulSoup(html2)
    time.sleep(random.uniform(3,6))
    
    html2 = html2.replace('<br>','<br><br>')
    
    ls1 = re.findall('</div>\n(.*?)<br>',html2)
    ls2 = re.findall('<br>\n(.*?)<br>',html2)
    ls3 = re.findall('<br>\n(.*?)</section>',html2)
    ls4 = re.findall('</div>\n(.*?)</section>',html2)
    
    if ls1 == [] and ls2 == [] and ls3 == [] and ls4 == []:
        prefinal = l
    elif ls4 == []:
        prefinal = ls1 + ls2 + ls3
    else:
        prefinal = ls4
    text.append(prefinal) 
    
    

    try:
        if soup.find('span', id = 'titletextonly').next_sibling.text[1] == '(' :
            title.append(soup.find('span', id = 'titletextonly').text + soup.find('span', id = 'titletextonly').next_sibling.text)
            
        else:
            title.append(soup.find('span', id = 'titletextonly').text)
            
    except:
        title.append(soup.find('span', id = 'titletextonly').text)
        continue
```

```python
# Write to CSV
import csv
with open('C:/Users/17654/Desktop/Fall Module 2/AUD/Project/Test.csv', 'w', encoding = 'utf-8', newline = '') as csvfile:
    
    fieldnames = ['S.No', 'URL', 'Title', 'Description']
    writer = csv.DictWriter(csvfile, fieldnames = fieldnames)
    
    writer.writeheader()
    for i in range (0,len(title)):
        writer.writerow({'S.No' : i, 'URL' : urls[i], 'Title' : title[i], 'Description' : text[i]})
```
